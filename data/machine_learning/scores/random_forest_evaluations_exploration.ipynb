```python
import pandas as pd
```

```python
df=pd.read_csv('random_forest_tt_scores.csv')
df1=pd.read_csv('random_forest_total_journey_tt_scores.csv')
df1=df1.rename(columns={"<60": "total<60", "<120": "total<120",">120": "total>120","overall": "total_overall"})
```

```python
df=pd.merge(df, df1, on=['route'])
```

```python
df['>120']=round(df['>120']/60, 1)
df['<120']=round(df['<120']/60, 1)
df['<60']=round(df['<60']/60, 1)
df['<30']=round(df['<30']/60, 1)
df['<10']=round(df['<10']/60, 1)
#df['<60']=round(df.sort_values('overall')['<60']/60, 2)
df['overall']=round(df['overall']/60, 1)
df['total<60']=round(df['total<60']/60, 1)
df['total<120']=round(df['total<120']/60, 1)
df['total>120']=round(df['total>120']/60, 1)
df['total_overall']=round(df['total_overall']/60, 1)
```
